```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Load CSV and preview DataFrame

equip_to_load= "Group Project Data.csv"

equip_df= pd.read_csv(equip_to_load)

equip_df.head()
```

```python
Types = equip_df["Type"].unique()
Types
```

```python
new_df = equip_df.dropna()
new_df
```

```python
Types = new_df["Type"].unique()
Types
```

```python
##DataFrame Clean-up
# Types
#index_type_df = new_df.set_index('Type')

updated_df = new_df[(new_df["Type"].str.contains("C")) | (new_df["Type"].str.contains("S"))]

updated_df
```

```python
updated_df["Type"].unique()
```

```python
updated_df
```

```python
grouped_df = updated_df.groupby(['Type'])
grouped_df
```

```python
total_cost = grouped_df["Total*"].sum()
total_cost.head()
```

```python
meter_max = grouped_df["Meter"].max()
meter_max
```

```python
meter_min = grouped_df["Meter"].min()
meter_min
```

```python
total_hours = meter_max - meter_min
total_hours
```

```python
cost_per_hour = total_cost/ total_hours
cost_per_hour
```

```python
CPH = pd.DataFrame({"Total Cost":total_cost,
                    "Total Meters": total_hours,
                     "Cost Per Hour":cost_per_hour})
CPH
```

```python
type_unique = updated_df['Type'].unique()
type_unique
```

```python
new_cph_df= pd.DataFrame ({"Type": type_unique,
                           "Cost":total_cost})
new_cph_df  
```

```python
Type = new_cph_df["Type"]
Cost = total_cost


plt.plot(Type,Cost)

# Orient widths. Add labels, tick marks, etc. 
plt.title("Total Cost by Type")
plt.ylabel("Total Cost")
plt.xlabel("Type")
#plt.xticks(np.arange(0,100000,5000))

# Save the Figure
plt.savefig("Total_Cost_by_Type_plot.png")

# Show the Figure
plt.show()
```

```python
Type = new_cph_df["Type"]
Cost = total_cost


plt.bar(Type,Cost)

# Orient widths. Add labels, tick marks, etc. 
plt.title("Total Cost by Type")
plt.ylabel("Total Cost")
plt.xlabel("Type")
#plt.xticks(np.arange(0,100000,5000))

# Save the Figure
plt.savefig("Total_Cost_by_Type_bar.png")

# Show the Figure
plt.show()
```

```python
Type = new_cph_df["Type"]
Cost_Hour= cost_per_hour

plt.bar(Type,Cost_Hour)

# Orient widths. Add labels, tick marks, etc. 
plt.title("Total Cost/ Hour by Type")
plt.ylabel("Total Cost")
plt.xlabel("Type")
#plt.xticks(np.arange(0,100000,5000))

# Save the Figure
plt.savefig("Total_Cost_Hour_by_Type_bar.png")

# Show the Figure
plt.show()
```
