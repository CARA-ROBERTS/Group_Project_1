```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Load CSV and preview DataFrame

equip_to_load= "Date_Group Project Data.csv"

equip_df= pd.read_csv(equip_to_load)

equip_df.head()
```

```python
Types = equip_df["Type"].unique()
Types
```

```python
new_df = equip_df.dropna()
new_df
```

```python
updated_df = new_df[(new_df["Type"].str.contains("C")) | (new_df["Type"].str.contains("S"))]

updated_df
```

```python
Types = updated_df["Type"].unique()
Types
```

```python
#CLARK, GENIE, HYSTE, JLG, KALMA, NISS, RICO, TAYLO, TOYO, YALE
updated_make_df = updated_df[(updated_df["Make"].str.contains("CLARK")) | (updated_df["Make"].str.contains("GENIE"))|
                             (updated_df["Make"].str.contains("HYSTE")) | (updated_df["Make"].str.contains("JLG"))|
                            (updated_df["Make"].str.contains("KALMA")) | (updated_df["Make"].str.contains("NISS"))|
                            (updated_df["Make"].str.contains("RICO")) | (updated_df["Make"].str.contains("TAYLO"))|
                            (updated_df["Make"].str.contains("TOYO")) | (updated_df["Make"].str.contains("YALE"))]

updated_make_df
```

```python
makes = updated_make_df["Make"].unique()
makes
```

```python
#& (updated_make_df["Meter"] ==9999) & (updated_make_df["Meter"] > 30000)
hour_munge_1 = updated_make_df[(updated_make_df["Meter"] ==999)].index
updated_make_df.drop(hour_munge_1, inplace = True)
```

```python
hour_munge_2 = updated_make_df[(updated_make_df["Meter"] ==9999)].index
updated_make_df.drop(hour_munge_2, inplace = True)
```

```python
hour_munge_3 = updated_make_df[(updated_make_df["Meter"] > 30000)].index
updated_make_df.drop(hour_munge_3, inplace = True)
```

```python
grouped_df = updated_make_df.groupby(["Make"])
grouped_df
```
